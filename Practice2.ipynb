```python
# shorting of matrix column wise
a=np.array([[2.1,1.2],[1.6,5.7]])
print(a)
a.sort(axis=0)
print(a)
```

```python
# shorting of matrix row wise
a=np.array([[2.1,1.2],[1.6,5.7]])
print(a)
a.sort(axis=1)
print(a)
```

```python
# Resize Array
a=np.array([[1,2,3,4],[6,8,9,10]])
a.resize(4,2) # (r,c)---r=row,c=column
print(a)
```

```python
# Append the array in flatten matrix (2 D Matrix)
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8,9],[10,11,12]])
c=np.append(a,b) #append operation of a nad b matrix and then flatten in a matrix(c)
print(a)
print(b)
print(c)
```

```python
# Flatten the array in 1d matrix
a=np.array([[1,2,3],[4,5,6]])
b=a.flatten()
print(a)
print(b)
```

```python
# Insert elemnt in the array
a=np.array([[1,2,3],[4,5,6]])
b=np.insert(a,2,10) #(a,i,e)--->a=array variable,i=index of the array in flatten,e=element which is inserted  
print(a)
print(b)
```

```python
# Concatenation the array in horizontal axis
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8,9],[10,11,12]])
c=np.concatenate((a,b),axis=1)
print(a)
print(b)
print(c)
```

```python
# Concatenation the array in vertical axis
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8,9],[10,11,12]])
c=np.concatenate((a,b),axis=0)
print(a)
print(b)
print(c)
```

```python
# Transpose of a matrix
a=np.array([[1,2,3],[4,5,6]])
b=a.T
print(a) #old matrix(a)
print(b) #new transpose matrix
```

```python
# Determinent of a matrix
a=np.array([[1,2],[4,5]])
b=np.linalg.det(a)
print(a) #old matrix(a)
print(b) #new determinent value of matrix (a)
```

```python
# Inverse of a matrix
a=np.array([[1,2],[4,5]])
b=np.linalg.inv(a)
print(a) #old matrix(a)
print(b) #Inverse of matrix (a)
```

```python
# random choice
data=[]
a=["a","b","c","d","e"]
for i in range(1000):
    value=np.random.choice(a,1,replace=False,p=[0.5,0.2,0.2,0.05,0.05])[0] # repetating value cannot be picking 
    data.append(value)

print(data)
```

```python
z=data.count("a")
y=data.count("b")
x=data.count("c")
w=data.count("d")
v=data.count("e")
u=len(data)
print("Count of picking a is :",z)
print("Count of picking b is :",y)
print("Count of picking c is :",x)
print("Count of picking d is :",w)
print("Count of picking e is :",v)
print("Total Count of element is :",u)
```
